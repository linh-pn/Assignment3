```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Create synthetic data
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Function to plot the K-means clustering
def plot_kmeans(X, kmeans, iteration):
    plt.figure(figsize=(8, 6))
    plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis', marker='o', edgecolor='k', s=50)
    plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', s=200, alpha=0.75, marker='X')
    plt.title(f"K-means Clustering - Iteration {iteration + 1}")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.grid()
    plt.show()

# Good initialization
k = 4  # Number of clusters
kmeans_good = KMeans(n_clusters=k, init='random', n_init=1, max_iter=1, random_state=42)

# Run K-means for 5 iterations
for i in range(5):
    kmeans_good.fit(X)  # Fit for one iteration
    plot_kmeans(X, kmeans_good, i)  # Plot the results after the iteration
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Create synthetic data
X, y = make_blobs(n_samples=500, centers=5, cluster_std=0.6, random_state=0)
df = pd.DataFrame(X, columns=['Feature 1', 'Feature 2'])

# Parameters
iterations = 5
centroids = None
k = 5

# Function to plot the K-means clustering
def plot_kmeans(X, kmeans, iteration):
    plt.figure(figsize=(8, 6))
    plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis', marker='o', edgecolor='k', s=50)
    plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', s=200, alpha=0.75, marker='X')
    plt.title(f"K-means Clustering - Iteration {iteration + 1}")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.grid()
    plt.show()

# K-means clustering with visualization
for i in range(iterations):
    kmeans = KMeans(
        max_iter=1,
        n_init=1,
        init=(centroids if centroids is not None else 'random'),
        n_clusters=k,
        random_state=1)

    kmeans.fit(X)
    centroids = kmeans.cluster_centers_

    print(f'iter: {i} - first: {centroids[0]}, second: {centroids[1]}')

    # Plot after each iteration
    plot_kmeans(X, kmeans, i)
```
