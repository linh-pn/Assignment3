```python
from sklearn.datasets import make_blobs
import pandas as pd

X, y = make_blobs(n_samples=1000, centers=5, cluster_std=0.60, random_state=0)
df = pd.DataFrame(X, columns=['Feature 1', 'Feature 2'])
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
k = 5  # Set the number of clusters
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(X)
```

```python
def plot_kmeans(X, kmeans, iteration):
    plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis', marker='o', edgecolor='k', s=50)
    plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', s=200, alpha=0.75, marker='X')
    plt.title(f"K-means Clustering - Iteration {iteration}")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.show()
```

```python
# Good initialization
kmeans_good = KMeans(n_clusters=k, init='random', max_iter=10, random_state=42)
kmeans_good.fit(X)
for i in range(5):
    plot_kmeans(X, kmeans_good, i)

# Bad initialization
kmeans_bad = KMeans(n_clusters=k, init=np.random.rand(k, 2), max_iter=10)
kmeans_bad.fit(X)
for i in range(5):
    plot_kmeans(X, kmeans_bad, i)
```
